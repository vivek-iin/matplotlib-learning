# Subplots

```python
# A diff way to plot graphs
batters.head()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Somthing')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
plt.show()
```

```python
# Drow two scatter plot
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')

plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(15,6))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])



plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1) #(2,1,1 where 2 show rows, 1 show column and anothe 1 show graph)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
plt.show()
```

# Surface Plots

```python

x = np.linspace(-10, 10, 100) 
y = np.linspace(-10, 10, 100)  
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python

# Generate values for  (in radians)
theta = np.linspace(0, 2 * np.pi, 500)

# Compute sin^2()
y = np.sin(theta)**2

# Plot the graph
plt.figure(figsize=(8, 4))
plt.plot(theta, y, label=r'$\sin^2(\theta)$', color='blue')

# Add labels, title, and grid
plt.title(r'Graph of $\sin^2(\theta)$')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$\sin^2(\theta)$')
plt.grid(True)
plt.legend()
plt.show()
```

# Contore Plots

```python
# contore plots means repsentation of 3d graph to 2d graph 
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Heatmap

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
# (delivery['ballnumber']).unique()
df=delivery[(delivery['ballnumber']).isin([1,2,3,4,5,6]) & (delivery['batsman_run']==6)]
```

```python
grid=df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')  # Save as a PNG file
```

```python
# Pie -> single and multiple
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)
df.head()
```

```python
# Ensure the length of `explode` matches the length of `df['batsman'].values`
explode = [0.1 if i == 3 else 0 for i in range(len(df['batsman']))]

df['match1'].plot(
    kind='pie',
    labels=df['batsman'].values,
    autopct='%0.1f%%',
    explode=explode
)
```

```python
# Multiple pie charts
df[['match1','match2','match3']].plot(kind='pie',subplots=True,figsize=(15,8))
```
