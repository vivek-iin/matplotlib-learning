```python
#from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#ploting multiple plots (campare kar rahe hai RG sharma aur virat ke performance ko)
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# lable title 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#color(hex) and line (width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='D',markersize='10')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#legend location (give lable to line)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='best') # best , upper right , upper left ,lower left ,lower right, right, center right, center left, lower center
```

```python
# limiting axes
price =[48000,54000,57000,47000,62000,999999]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# size
tips = sns.load_dataset('tips')

# slower technique 
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# Scatterplot using plt.plot

# faster  technique 
plt.plot(tips['total_bill'],tips['tip'],'o')
```

# Bar chat

```python
# simple bar chat
children = [10,20,30,50,89]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='yellow')
```

```python
# horizontal bar chat
plt.barh(colors,children,color='yellow')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df.head(5)
```

```python
x = np.arange(df.shape[0])  # x positions for bars
plt.bar(x - 0.2, df['2015'], width=0.2, label='2015')
plt.bar(x, df['2016'], width=0.2, label='2016')
plt.bar(x + 0.2, df['2017'], width=0.2, label='2017')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart ')
plt.legend()

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show()
```

```python
# problem (jo mujhe futer me aayega)
children = [10,20,30,50,89]
colors = ['red red red red red red red','blue lorem','green red red red rde rder red erd ','yellow','pink']

plt.bar(colors,children,color='yellow')

plt.xticks(rotation='vertical')

plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=df['2017'], label='2016')
plt.bar(df['batsman'], df['2015'], bottom=(df['2017'] + df['2016']), label='2015')

# Add labels, legend, and title
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Stacked Bar Chart of Runs by Year')
plt.legend()
plt.show()
```

# Pie chart

```python
# simple data
data = [23,45,100,20,49]
subjects= ['eng','hin','sci','maths','sst']

plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%', colors=['blue','black','green','red','pink'])
plt.show()
```

```python
# explor shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0.3,0,0,0,0.1],shadow=True)
plt.show()
```
